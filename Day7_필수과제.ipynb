- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
nums = []
while True:
    try:
        given_nums = input("숫자를 콤마로 구분해 입력하세요. 종료: done")
        if given_nums == "done":
            break
        nums.append(int(given_nums))

    except ValueError :
        print(f'잘못된 값:{given_nums}. 숫자를 입력해주세요.')

    finally:
        print("리스트 저장을 종료합니다.")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
#ai로 작성.. 다시 풀어보기
import time

# 1. 데코레이터 함수 정의
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        # [함수 실행 전] 시간 기록
        start_time = time.time()
        
        # 실제 함수 실행
        result = func(*args, **kwargs)
        
        # [함수 실행 후] 시간 기록
        end_time = time.time()
        
        # 실행 시간 계산 및 출력
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
        
        return result
    return wrapper

# 2. 데코레이터 사용하기
@timer_decorator
def test_function():
    print("함수가 실행 중입니다...")
    time.sleep(2)  # 2초 동안 프로그램을 멈춤 (테스트용)

# 3. 함수 실행
test_function()
````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
#ai로 작성.. 다시 풀어보기
# mymath.py
import math  # 원주율(pi) 등을 사용하기 위해 불러옵니다.

# 1. 삼각형의 넓이 구하는 함수
def triangle_area(base, height):
    return (base * height) / 2

# 2. 원의 넓이 구하는 함수
def circle_area(radius):
    # 원주율(3.14...) * 반지름의 제곱
    return math.pi * (radius ** 2)

# 3. 직육면체의 겉넓이 구하는 함수
def cuboid_surface_area(width, length, height):
    # (가로*세로 + 세로*높이 + 가로*높이) * 2
    return 2 * (width * length + length * height + width * height)
````

````python
# main.py
import mymath  # 우리가 만든 mymath.py를 불러옵니다.

# 1. 삼각형 넓이 테스트
tri_base = 10
tri_height = 5
print(f"삼각형 넓이 (밑변 {tri_base}, 높이 {tri_height}): {mymath.triangle_area(tri_base, tri_height)}")

# 2. 원의 넓이 테스트
radius = 7
# 결과가 너무 길면 소수점 둘째 자리까지 반올림(round)해서 출력해봅시다.
print(f"원 넓이 (반지름 {radius}): {round(mymath.circle_area(radius), 2)}")

# 3. 직육면체 겉넓이 테스트
w, l, h = 3, 4, 5
print(f"직육면체 겉넓이 (가로 {w}, 세로 {l}, 높이 {h}): {mymath.cuboid_surface_area(w, l, h)}")
````

- 각 모듈에는 최소한 한 개 이상의 동물에 관한 클래스를 정의하세요<br>
  (예: mammals에는 Dog, birds에는 Eagle)
- 이 패키지와 모듈을 사용하여 동물들의 정보를 출력하는 프로그램을 작성

폴더 및 파일 구조
```
animals/(폴더)
├── __init__.py
├── mammals.py
└── birds.py
```


````python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def info(self):
        return f"이름: {self.name}, 종류: 포유류(강아지), 품종: {self.breed}"

    def speak(self):
        return "멍멍!"
````

````python
class Eagle:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def info(self):
        return f"이름: {self.name}, 종류: 조류(독수리), 종: {self.species}"

    def speak(self):
        return "훨훨"
````
