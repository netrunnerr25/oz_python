[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_nums(n):
    for i in range(n + 1):
        if i % 2 != 0:
            yield i

for num in odd_nums(10):
    print(num)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max():
    n1 = int(input("첫번째 숫자:"))
    n2 = int(input("첫번째 숫자:"))
    n3 = int(input("첫번째 숫자:"))
    nums = (n1, n2, n3)
    print("입력받은 숫자 3개 :", nums)
    return min(nums), max(nums)

최소, 최대 = min_max()
print(f"최솟값: {최소}, 최댓값: {최대}")
```

```python
def is_even(num):
    if num % 2 == 0:
        return  True
    
    return False

result = is_even(5)
print("짝수인가요?", result)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

---

```python
def factorial(n: int) -> int: 
    #종료 조건
    if n <= 1:
        return 1
    #재귀 호출
    return n * factorial(n - 1)

print(factorial(5))
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(*args):
    for i in args:
        print("문자열 길이 :", len(i))

print_lengths("지구젤리", "두쫀쿠", "두바이초콜렛")
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def long_str(str1: str, str2: str) -> str:
    if len(str1) >= len(str2):
        return str1
    return str2 #조기 리턴

print(long_str("두쫀쿠", "두바이 초콜렛"))
print(long_str("지구젤리", "두쫀쿠"))
```
